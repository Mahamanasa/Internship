Q1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: A)
Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#import necessary libraries
import pandas as pd
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.by import By
```

```python
#connecting chrome webdriver
driver=webdriver.Chrome()
driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")

#scroll down the page
driver.execute_script("window.scrollTo(0, 500)")  #500 is the height that needs to be scrolled
```

```python
#Finding Details
rows = driver.find_elements(By.XPATH, '/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[1]/tbody/tr')
details = []
ranks=[]
name=[]
artist=[]
views=[]
upload=[]
date=[]
rank = 1
for row in rows:
    cells = row.find_elements(By.XPATH, 'td')
    ranks.append(rank)
    name.append(cells[0].text)
    artist.append(cells[1].text)
    views.append(cells[2].text)
    date.append(cells[3].text)
    rank += 1

print(len(ranks),len(name),len(artist),len(views))
```

```python
#creating dataframe
df=pd.DataFrame({})
df["Rank"]=ranks
df["Name"]=name
df["Artist"]=artist
df['Upload date']=date
df['Views']=views
df
```

Q4.Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
#get the URL
driver=webdriver.Chrome()
driver.get("https://github.com/trending")
```

```python
#trending=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/header/div/div[2]/div/nav/ul/li[4]/div/div[3]/ul/li[2]/a')
#trending.click()
```

```python
#creating empty list
repository_title=[]
repository_description=[]
contributors_count=[]
language_used=[]
#title
title_tag=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in title_tag:
    repository_title.append(i.text)
    
#url list
url_list=[]
url=driver.find_elements(By.XPATH,'//h2[@class="h3 lh-condensed"]/a')
for i in url:
    url_list.append(i.get_attribute("href"))
    
    
```

```python
#visit each repository
for i in url_list:
    driver.get(i)
    #repository description
    try:
        description=driver.find_element(By.XPATH,"//div[@class='BorderGrid-cell']/p")
        repository_description.append(description.text)
    except NoSuchElementException as e:
        repository_description.append("---")
     #contributors count  #use contains in xpath to fetch only contributors
    try:
        count=driver.find_element(By.XPATH,"//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[5]/div/h2/a[contains(text(),'Contributors ')]")
        contributors_count.append(count.text.split(" ")[1])
    except NoSuchElementException as e:
        contributors_count.append("---")
    #language used
    try:
        lang=driver.find_element(By.XPATH,"//*[@id='repo-content-pjax-container']/div/div[2]/div[2]/div/div[6]/div/ul")
        language_used.append(lang.text.replace("\n"," "))
    except NoSuchElementException as e:
        language_used.append("---")
        
```

```python
print(len(repository_title),len(repository_description),len(contributors_count),len(language_used))
```

```python
#creating dataset
df=pd.DataFrame({"Repository Title":repository_title,"Repository Desc":repository_description,"Count":contributors_count,"Language":language_used})
df
```

have to find the following details:

C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year

```python
#fetching URL
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
```

```python
#click on view all datasets
#dataset=driver.find_elements(By.XPATH,"//a[@href='datasets.php']")
datasets = driver.find_element(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]')
print(datasets)
datasets.click()

```

```python
# Initialize lists to store dataset details
dataset_names = []
data_types = []
tasks = []
attribute_types = []
num_instances = []
num_attributes = []
years = []

name=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[1]/div[2]/h2/a')
for i in name:
    dataset_names.append(i.text)
                        
type=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[1]/div[2]/div/div[1]/span')
for i in type:
    data_types.append(i.text)
    
task=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[1]/div[2]/div/div[2]/span')
for i in task:
    tasks.append(i.text)

attribute=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[2]/div/table/tbody/tr/td[2]')
for i in attribute:
    attribute_types.append(i.text)
    
instance=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[3]/span')
for i in instance:
    num_instances.append(i.text)
    
attribute=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div/div[2]/div/div[4]/span')
for i in attribute:
    num_attributes.append(i.text)
    
year=driver.find_elements(By.XPATH,'/html/body/div/div[1]/div[1]/main/div/div[2]/div[2]/div[1]/div[2]/div/table/tbody/tr/td[3]')
for i in year:
    years.append(i.text)
    

                          
                          
```

```python
#checking the lengths
print(len(dataset_names))
print(len(data_types))
print(len(tasks))
print(len(attribute_types))
print(len(num_attributes))
print(len(num_instances))
print(len(years))
```
