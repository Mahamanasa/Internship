```python
#Q3: Scrape 100 reviews data from flipkart.com for iphone11 phone.
#Initializing the website
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")
```

```python
#Scraping out rating, review summary and full review data from the site
rating=[]
review_summary=[]
full_review=[]
start=0
end=10

for page in range(start,end):
    rating_tags=driver.find_elements(By.XPATH,'//div[@class="XQDdHH Ga3i8K"]')
    for i in rating_tags:
        ratings=i.text
        rating.append(ratings)
        
    review_summary_tags=driver.find_elements(By.XPATH,'//p[@class="z9E0IG"]')
    for i in review_summary_tags:
        rev_sum=i.text
        review_summary.append(rev_sum)
        
    full_review_tags=driver.find_elements(By.XPATH,'//div[@class="ZmyHeo"]')
    for i in full_review_tags:
        full_rev=i.text
        full_review.append(full_rev)
        
    next_button=driver.find_element(By.CLASS_NAME,"_9QVEpD")
    next_button.click()
    time.sleep(2)
    
```

```python
print(len(rating),len(review_summary),len(full_review))
```

```python
#Creating the dataset for the scraped data.
apple_iphone_df=pd.DataFrame({"Ratings":rating,"Review Summary":review_summary,"Full Review":full_review})
apple_iphone_df
```

```python
#Q4: Scrape data for first 100 sneakers you find when you visit flipkart.com and search for ‚Äúsneakers‚Äù in the search
#field.
#Initializing the website
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/")
```

```python
#Scraping data with the input criteria as Sneakers
product_title=driver.find_element(By.CLASS_NAME,'zDPmFV')
product_title.send_keys("sneakers")
```

```python
#Initializing the search button
search=driver.find_element(By.CLASS_NAME,'MJG8Up')
search.click()
```

```python
#Scraping out brand,product description and price from the searched data output
brand=[]
product_desc=[]
price=[]
start=0
end=3

for page in range(start,end):
    brand_tags=driver.find_elements(By.XPATH,'//div[@class="syl9yP"]')
    for i in brand_tags:
        brands=i.text
        brand.append(brands)
   
        
    product_desc_tags=driver.find_elements(By.XPATH,'//a[@class="WKTcLC"]')
    product_tags=driver.find_elements(By.XPATH,'//a[@class="WKTcLC BwBZTg"]')
    for i in product_desc_tags:
        desc_tag=i.text
        product_desc.append(desc_tag)
    for i in product_tags:
        desc=i.text
        product_desc.append(desc)
   
        
    price_tags=driver.find_elements(By.XPATH,'//div[@class="Nx9bqj"]')
    for i in price_tags:
        prices=i.text
        price.append(prices)

      
    next_button=driver.find_element(By.CLASS_NAME,"_9QVEpD")
    next_button.click()
    time.sleep(2)
    
    
    
```

```python
brand_final=[]
product_final=[]
price_final=[]
brand_final=brand[:100]
product_final=product_desc[:100]
price_final=price[:100]
print(len(brand_final),len(product_final),len(price_final))
```

```python
#Creating the dataset for the scraped data
df=pd.DataFrame({"Brand":brand_final,"Product Description":product_final,"Price":price_final})
df
```

```python
#Q6: Write a python program to scrape data for Top 1000 Quotes of All Time.
#Initializing the site
driver=webdriver.Chrome()
driver.get("https://www.azquotes.com/")
```

```python
#Setting the search criteria as top quotes
top_quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div/div[3]/ul/li[5]/a')
top_quotes.click()
```

```python
#Scraping out the quote, quote type and the author from the top quotes criteria -1000 Quotes
quote=[]
author=[]
type_quote=[]
start=0
end=10
for page in range(start,end):
    quote_tag=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quote_tag:
        quotes=i.text
        quote.append(quotes)
        
    author_tag=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in author_tag:
        authors=i.text
        author.append(authors)
        
    type_tag=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in type_tag:    
        types=i.text
        type_quote.append(types)
        
    next_button=driver.find_element(By.CLASS_NAME,"next")
    next_button.click()
    time.sleep(1)
    
```

```python
print(len(quote),len(author),len(type_quote))
```

```python
quotes_df=pd.DataFrame({"Quote":quote,"Author":author,"Quote Type":type_quote})
quotes_df
```
