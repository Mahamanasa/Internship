```python
#3. Write a python program to access the search bar and search button on images.google.com and scrape 10
#images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’.
#importing necessary libraries
import urllib
import os
import shutil
import base64
from selenium.webdriver.common.keys import Keys

#funtion for calling on the url
def scrape_images_for_links(keywords, image_limit=10):
    driver = webdriver.Chrome()
    url = "https://images.google.com/"
    driver.get(url=url)    


try:
    reject_all_button = driver.find_element(By.XPATH, "//button[.//div[text()='Reject all']]")
    reject_all_button.click()
except Exception as e:
    print("Shouldn't happen", e)
        
keyword_to_src_links = {}
for keyword in keywords:
    textarea = driver.find_element(By.XPATH, "//textarea")
    textarea.click()
    textarea.send_keys(keyword)
    textarea.send_keys(Keys.ENTER)
    time.sleep(1)
    images = driver.find_elements(By.XPATH, "//img[contains(@class,'YQ4gaf')]")
    keyword_to_src_links[keyword] = []
        
for image in images:
    if (image.get_attribute("src").startswith("data:image/jpeg;base64")):
        keyword_to_src_links[keyword].append(image.get_attribute("src"))
    if len(keyword_to_src_links[keyword]) == image_limit:
        break
                
driver.back()
driver.close()
return keyword_to_src_links

def download_images(keyword_to_src_links):
    for keyword, links in keyword_to_src_links.items():
        current_directory = os.getcwd()
        folder_path = os.path.join(current_directory, keyword)
        if os.path.exists(folder_path):
            shutil.rmtree(folder_path)
        os.mkdir(folder_path)
        cnt = 1
        for link in links:
            base64_image_data = link.split(",")[1]
            download_image(base64_image_data, folder_path + "/image" + str(cnt) + ".jpeg")
            cnt += 1
            
def download_image(base64_image_data, save_path):
    image_data = base64.b64decode(base64_image_data)
    with open(save_path, 'wb') as f:
        f.write(image_data)


def scrape_images_and_download(keywords, image_limit):
    keyword_to_src_links = scrape_images_for_links(keywords, image_limit)
    download_images(keyword_to_src_links)


warnings.filterwarnings('ignore')
keywords_to_scrape = ["fruits", "cars", "Machine Learning", "Guitar", "Cakes"]

scrape_images_and_download(keywords_to_scrape, 10)

    
    
```

```python
#7. Write a python program to scrape the details for all billionaires from www.forbes.com. Details to be scrapped:
#“Rank”, “Name”, “Net worth”, “Age”, “Citizenship”, “Source”, “Industry”.
#import necessary libraries

import time

from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys

def scrape_page(content, billionaires):
    soup = BeautifulSoup(content, 'html.parser')
    rows = soup.find_all('div', class_='Table_tableRow__lF_cY')
    for row_div in rows:
        name = row_div.find_all('div', role='cell')[1].text.strip()
        net_worth = row_div.find_all('div', role='cell')[2].text.strip()
        age = row_div.find_all('div', role='cell')[3].text.strip()
        country = row_div.find_all('div', role='cell')[4].text.strip()
        source = row_div.find_all('div', role='cell')[5].text.strip()
        industry = row_div.find_all('div', role='cell')[6].text.strip()
        billionaires.append({
            name: name,
            net_worth: net_worth,
            age: age,
            country: country,
            source: source,
            industry: industry
        })
    return billionaires


def billionaires_from_forbes():
    driver = webdriver.Chrome()
    driver.set_window_size(1920, 1080)  # important since smaller window cuts down information
    url = "https://www.forbes.com/billionaires/"
    driver.get(url=url)
    time.sleep(5)
    try:
        driver.find_element(By.XPATH, './/button[@aria-label="Accept All"]').click()
    except Exception as e:
        print("Unexpected exception", e)
    time.sleep(5)
    billionaires = []
    page_num = 1
    driver.find_element(By.XPATH, '//body').send_keys(Keys.ESCAPE)

    while True:
        print("Page number: " + str(page_num))
        time.sleep(1)  # give it time to load
        billionaires = scrape_page(driver.page_source, billionaires)

        # go to next page
        next_button = driver.find_element(By.CSS_SELECTOR, 'button[aria-label="Next"]')
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        page_num += 1
        if next_button.is_enabled():
            driver.execute_script("arguments[0].click();", next_button)
        else:
            break
    return billionaires


print(billionaires_from_forbes())

```

```python
#8. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted
#from any YouTube Video.

#importing libraries
    

import time

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys


COMMENT_COUNT_TO_EXTRACT = 100  # change this to 500 when you run the code


def extract_comments_from_youtube_video(url, count):
    driver = webdriver.Chrome()
    driver.set_window_size(1920, 1080)
    driver.get(url=url)
    time.sleep(5)
    accept_all = driver.find_element(By.XPATH, './/button[@aria-label="Accept the use of cookies and other data for the purposes described"]')
    driver.execute_script("arguments[0].click();", accept_all)
    time.sleep(1)
    while len(driver.find_elements(By.XPATH, '//ytd-comment-view-model')) < count:
        driver.find_element(By.XPATH, '//body').send_keys(Keys.PAGE_DOWN)
        time.sleep(0.5)
    comments = []
    for comment_model in driver.find_elements(By.XPATH, '//ytd-comment-view-model')[:count]:
        soup = BeautifulSoup(comment_model.get_attribute("outerHTML"), 'html.parser')
        comment_text = soup.find('yt-attributed-string', {'id': 'content-text'}).text.strip()
        posted_time = soup.find('span', {'id': 'published-time-text'}).text.strip()

        try:
            upvotes_count = soup.find('span', {'id': 'vote-count-middle'}).text.strip()
        except Exception as e:
            upvotes_count = '0'

        comments.append({
            'text': comment_text,
            'posted_time': posted_time,
            'upvotes_count': upvotes_count
        })

    return comments


# Stay on the page fully, YouTube ensures you're on the page when it displays cookie popups.
extracted_comments = extract_comments_from_youtube_video(
    url="https://www.youtube.com/watch?v=qhW1HfSuPVQ",
    count=COMMENT_COUNT_TO_EXTRACT
)

print(extracted_comments)
   
                                                
       

                                                
```
