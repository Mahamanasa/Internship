```python
#4. Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com
#and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand
#Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”,
#“Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the
#details is missing then replace it by “- “. Save your results in a dataframe and CSV

#fetching URL
driver=webdriver.Chrome()
driver.get("https://www.flipkart.com/")

```

```python
#search the search product "smartphone"
search_bar=driver.find_element(By.XPATH,"/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input")
search_bar.send_keys("smartphone")
```

```python
#click on search
search_button=driver.find_element(By.XPATH,"/html/body/div/div/div[1]/div[1]/div[2]/div[2]/form/div/button")
search_button.click()
```

```python
#grab all the links of mobile phones to visit to next page 
url_list=[]
for i in driver.find_elements(By.XPATH,"//a[@class='CGtC98']"):
    url_list.append(i.get_attribute("href"))
url_list
```

```python
#create empty lists
brand_list=[]
smartphone=[]
color_list=[]
ram_list=[]
rom_list=[]
primary_camera_list=[]
secondary_camera_list=[]
display_list=[]
battery_list=[]
price_list=[]
product_url=[]
```

```python
#go to the link of each mobile phone and extract the details required
for a in url_list:
    driver.get(a)
    #brand (as the brand is not mentioned on web page we will scrap phone name only as brand)
    brand=driver.find_element(By.XPATH,"//span[@class='VU-ZEz']")
    brand_list.append(brand.text)
    
     #smartphone name
    title=driver.find_element(By.XPATH,"//span[@class='VU-ZEz']")
    smartphone.append(title.text)
    
    #click on read more button to get all specifications
   # try:
    #read_more=driver.find_element(By.XPATH,'//html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/button')
  #  except NoSuchElementException as e:
    #read_more=driver.find_element(By.CLASS_NAME,"QqFHMw _4FgsLt")
 #  read_more.click()
                          
    #color
    color=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[1]/table/tbody/tr[4]/td[2]/ul/li')
    color_list.append(color.text)
   
    #ram
    try:
        ram=driver.find_element(By.XPATH,'//html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[2]/td[2]/ul/li')
        ram_list.append(ram.text) 
    except NoSuchElementException as e:
        ram_list.append("--") 
                              
    #storage/rom
    try:
        rom=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[4]/table/tbody/tr[1]/td[2]/ul/li')
        rom_list.append(rom.text)
    except NoSuchElementException as e:
        rom_list.append("--")
                                
    #primary camera (handling exception)
    try:
        pcam=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[2]/td[2]/ul/li')
        primary_camera_list.append(pcam.text)
    except NoSuchElementException as e:
        primary_camera_list.append("--")
                              
    #secondary camera (handling exception)
    try:
        
        sec_cam=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')  
        secondary_camera_list.append(sec_cam.text)
    except NoSuchElementException as e:
        secondary_camera_list.append("--") 
    
    #display
    try:
        display=driver.find_element(By.XPATH,"/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[2]/table/tbody/tr[1]/td[2]/ul/li")
        display_list.append(display.text) 
    except NoSuchElementException as e:
        display_list.append("--")
                                 
    #battery
    try:
        battery=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[9]/div[5]/div/div[2]/div[1]/div[10]/table/tbody/tr/td[2]/ul/li')
        battery_list.append(battery.text) 
    except NoSuchElementException as e:
        battery_list.append("--") 
                                 
    #price
    price=driver.find_element(By.XPATH,'//div[@class="Nx9bqj CxhGGd"]')
    price_list.append(price.text)
    
    #product url
    img=driver.find_element(By.XPATH,'//div[@class="_4WELSP"]/img')
    product_url.append(img.get_attribute("src")) 
    
```

```python
print(len(brand_list),len(smartphone),len(color_list),len(ram_list),len(rom_list),len(primary_camera_list))
print(len(secondary_camera_list),len(display_list),len(battery_list),len(price_list),len(product_url))
```

```python
brand_name=brand_list[:30]
name=smartphone[:30]
color=color_list[:30]
ram=ram_list[:30]
rom=rom_list[:30]
prim_cam=primary_camera_list[:30]
second_cam=secondary_camera_list[:30]
disp=display_list[:30]
battery_spec=battery_list[:30]
rate=price_list[:30]
image_prod=product_url[:30]
```

```python
#creating the dataset
df=pd.DataFrame({"Brand Name":brand_name,"Title":name,"Color":color,"RAM":ram,"ROM":rom,"Primary Camera":prim_cam,"Secondary Camera":second_cam,"Display":disp,"Battery":battery_spec,"Price":rate,"Image":image_prod})
df
```

```python
#creating csv file
df.to_csv("Smartphones.csv")
```

```python
review_rating=[]
try:
    for i in driver.find_elements(By.XPATH,'//div[@class="rh_woo_star"]'):
        review_rating.append(i.text)
    
except NoSuchElementException as e:
    review_rating.append('--')

print(review_rating)
```

```python
price=[]
for i in driver.find_elements(By.XPATH,'//span[@class="woocommerce-Price-amount amount"]'):
    price.append(i.text)
    
print(price)
```

```python
df=pd.DataFrame({})
df['Title']=title
df['Review Rating']=review_rating
df['Price']=price
df
```
